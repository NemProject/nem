```python
import numpy as np
import scipy.stats
import scipy
import scikits.bootstrap as bootstrap
import seaborn
from scipy.stats import rankdata
np.set_printoptions(suppress=True)
```

```python
with open ("importances.csv", "r") as myfile:
    importances=eval(myfile.read().replace('\n', ''))
```

```python
with open ("stakes.csv", "r") as myfile:
    stakes=eval(myfile.read().replace('\n', ''))
```

```python
with open ("addresses.csv", "r") as myfile:
    addresses=eval(myfile.read().replace('\n', ''))
```

```python
with open ("outlinkCount.csv", "r") as myfile:
    outlinkCounts=eval(myfile.read().replace('\n', ''))
```

```python
print 'average outlink count', np.mean(outlinkCounts)
print 'median outlink count', np.median(outlinkCounts)
print '.95 CI around mean of outlink count',  bootstrap.ci(data=outlinkCounts, statfunction=scipy.mean)
```

```python
scipy.stats.pearsonr(importances, stakes)
```

```python
scipy.stats.pearsonr(stakes, outlinkCounts)
```

```python
scipy.stats.pearsonr(importances, outlinkCounts)
```

```python
# normalize to between 0 and 1
stakeSum = np.sum(stakes)*1.
stakesNorm = [i/stakeSum for i in stakes]

outlinkCountSum = np.sum(outlinkCounts)*1.
outlinkCountNorm = [i/outlinkCountSum for i in outlinkCounts]
```

```python
plot(outlinkCountNorm)
plot(importances)
plot(stakesNorm)
legend(["outlink #", 'importances', 'stakes'])
# yscale('log')
```

```python
rankedStakes = rankdata(stakesNorm, method="dense")
rankedImportances = rankdata(importances, method="dense")
rankedOutlinkCounts = rankdata(outlinkCountNorm, method="dense")
```

```python
plot(rankedOutlinkCounts[:30])
plot(rankedImportances[:30])
plot(rankedStakes[:30])
legend(['outlink #', 'rankedImportances', 'rankedStakes'])
```
